```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
from sklearn.cluster import KMeans
```

```python
raw_data = pd.read_csv('Countries_exercise.csv')
raw_data
```

```python
data = raw_data.copy()
```

```python
plt.scatter(data['Latitude'], data['Longitude'])
```

```python
x = data.iloc[:,1:3]
x
```

```python
kmeans = KMeans(4)
kmeans.fit(x)
```

```python
identified_clusters = kmeans.fit_predict(x)
identified_clusters
```

```python
dataWithClusters = data.copy()
dataWithClusters['Clusters'] = identified_clusters
dataWithClusters
```

```python
plt.scatter(dataWithClusters['Latitude'], dataWithClusters['Longitude'], c=dataWithClusters['Clusters'], cmap='rainbow')
plt.show()
```

```python
kmeans.inertia_
```

```python
wcss = []
cl_num = 11
for i in range(1, cl_num):
    kmeans = KMeans(i)
    kmeans.fit(x)
    wcss_iter = kmeans.inertia_
    wcss.append(wcss_iter)
```

```python
wcss
```

```python
numberClusters = range(1,cl_num)
plt.plot(numberClusters, wcss)
plt.show
```
